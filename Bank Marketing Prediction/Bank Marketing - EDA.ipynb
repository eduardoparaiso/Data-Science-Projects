# Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
sns.pairplot(data = df, hue = 'y')
```

```python
targets = ['age', 'education', 'campaign', 'pdays', 'previous']
```

## By age

```python
for target in targets:
    sns.countplot(x = target, data = df)
    plt.show()
```

```python
for target in targets:
    sns.scatterplot(x = target, y = 'age', hue = 'y', data = df)
    plt.show()
```

```python
for target in targets:
    sns.lineplot(x = target, y = 'age', hue = 'y', data = df)
    plt.show()
```
