## Looking for missing values

```python
#Checks whether the purchase IP is valid or non-standard
def isp_counts(isp):
    country = ISP_df[(ISP_df.limite_inferior_ip <= isp) & (ISP_df.limite_superior_ip >= isp)].pais
    if (country.shape[0] == 1):
        return print('Unique')
    elif (country.shape[0] > 1):
        return print('Many countries')
    elif (country.shape[0] == 0):
        if (isp > ISP_df.limite_superior_ip.max()):
            return print ('Probable Fraud') #Too big
        elif (isp < ISP_df.limite_inferior_ip.min()):
            return print ('Probable Fraud') #Too Small
        else:
            return print ('Missing')
        
ip_counts = fraud_df.ip.apply(isp_counts)
```

```python
#Reads all IPs and associates them with its country
def ip_country(ip):
    tmp = ISP_df[(ISP_df.limite_inferior_ip <= ip) & (ISP_df.limite_superior_ip >= ip)].pais
    if (tmp.shape[0] == 1):
        return tmp.iloc[0]
    elif (tmp.shape[0] < 1):
        return print('Missing')
    elif (tmp.shape[0] > 1):
        return print('ERROR')
    
fraud_df['country'] = fraud_df.ip.apply(ip_country)
```
