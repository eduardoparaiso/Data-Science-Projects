# Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(x = 'Age', kde = True, hue= 'class', data = df)
plt.show()
```

```python
sns.countplot(x = 'Gender', data = df)
plt.show
```

```python
sns.countplot(x = 'Gender', hue = 'class', data = df)
plt.show()
```

```python
symptoms = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia', 'Genital thrush', 'visual blurring',
            'Itching', 'Irritability', 'delayed healing', 'partial paresis', 'muscle stiffness', 'Alopecia', 'Obesity']
```

```python
for symptom in symptoms:
    sns.countplot(x = symptom, hue = 'class', data = df)
    plt.show()
```

```python
for symptom in symptoms:
    sns.violinplot(x = symptom, y = 'Age', hue = 'class', data = df)
    plt.show()
```

```python
for symptom in symptoms:
    sns.lineplot(x = symptom, y = 'Age', hue = 'class', data = df)
    plt.show()
```

```python
sns.countplot(x= 'class', data = df)
plt.show()
```
